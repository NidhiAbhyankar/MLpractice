```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
movie_attributes=pd.read_csv('movies_metadata.csv')
print(movie_attributes.shape)
movie_attributes = movie_attributes[:5000]
movie_attributes.head()
```

```python
def get_title_from_index(index):
	return movie_attributes[movie_attributes.index==index]["title"].values[0]

def get_index_from_title(title):
	return movie_attributes[movie_attributes.title==title]["index"].values[0]
```

```python
movie_attributes['new_col']=range(0,len(movie_attributes))
movie_attributes=movie_attributes.reset_index()
movie_attributes.head()
```

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.metrics.pairwise import cosine_similarity




features=['genres','overview','tagline']
#create a column in df which combines all selected features
for feature in features:
	movie_attributes[feature]=movie_attributes[feature].fillna('')

def combine_features(row):
	try:
		return row['genres']+" "+row['overview']+" "+row['tagline']
	except:
		print("error",row)

movie_attributes["combine_features"]=movie_attributes.apply(combine_features,axis=1) #each row individually
print("Combined features: ",movie_attributes["combine_features"].head())


#create count matrix from this new combined column
cv=CountVectorizer()
count_matrix=cv.fit_transform(movie_attributes["combine_features"])
```

```python
#compute the cosine similarity based on the count_matrix
import csv
cosine_sim=cosine_similarity(count_matrix)

print(type(cosine_sim))
j=1
while j:
    #print("Enter Movie for which you need recommendation.")
    movie_user_likes=input()
    

    #get the index of this movie from its title
    movie_index=get_index_from_title(movie_user_likes)
    similar_movies=list(enumerate(cosine_sim[movie_index]))

    #get a list of similar movies in descending order
    sorted_similar_movies=sorted(similar_movies,key=lambda x:x[1],reverse=True)[1:]
    i=0
    for movie in sorted_similar_movies:
        print(get_title_from_index(movie[0]))
        i=i+1
        if(i>10):
            break
    print("Want to check another movie?(y/n)")
    choice=input().upper()
    if(choice!='Y'):
        j=0
   
```
